```python
import numpy as np
import matplotlib.pyplot as plt
import time
from sklearn.decomposition import DictionaryLearning

# ==========================================
# 1) SIMULACIÓN DEL SENSOR INDUSTRIAL (12.000 Hz)
# ==========================================
print("Generando señal de vibración de un motor con rodamiento dañado...")
fs = 12000  # Frecuencia de muestreo (12 kHz)
t = np.arange(0, 1.0, 1/fs)  # 1 segundo de grabación (12.000 puntos)

# A) Ruido normal del eje girando (Baja frecuencia)
eje_sano = 0.5 * np.sin(2 * np.pi * 30 * t) 

# B) Impactos del rodamiento roto (Alta frecuencia que decae)
frec_impacto = 10  # Golpea 10 veces por segundo
impactos = np.zeros_like(t)
impactos[::int(fs/frec_impacto)] = 1.0

# Creamos la onda de resonancia del metal (el "clack" agudo)
tiempo_res = np.arange(0, 0.02, 1/fs)
resonancia = np.exp(-300 * tiempo_res) * np.sin(2 * np.pi * 2500 * tiempo_res)

# Unimos todo y le sumamos ruido blanco de la fábrica
ruido_fabrica = np.random.normal(0, 0.3, len(t))
senal_vibracion = eje_sano + np.convolve(impactos, resonancia, mode='same') + ruido_fabrica

# ==========================================
# 2) PREPARACIÓN DE LA MATRIZ (VENTANEO)
# ==========================================
# Cortamos el segundo de audio en 100 trozos (ventanas) de 120 puntos cada uno
longitud_ventana = 120
num_ventanas = len(senal_vibracion) // longitud_ventana

# Nuestra matriz X: 100 ejemplos, 120 dimensiones físicas (milisegundos)
X_vibracion = senal_vibracion[:num_ventanas * longitud_ventana].reshape(num_ventanas, longitud_ventana)

print(f"Matriz acústica lista: {X_vibracion.shape[0]} ventanas con {X_vibracion.shape[1]} lecturas cada una.")
print("-" * 60)

# ==========================================
# 3) SPARSE CODING EXTREMO (TODOS LOS NÚCLEOS)
# ==========================================
n_atoms = 15
lambda_param = 1.0

print(f"Extrayendo {n_atoms} firmas acústicas usando el 100% de la CPU...")
inicio_tiempo = time.time()

dict_learner = DictionaryLearning(
    n_components=n_atoms, 
    alpha=lambda_param, 
    transform_algorithm='lasso_lars', 
    random_state=42,
    max_iter=1000,
    n_jobs=-1  # ¡MÁXIMA POTENCIA DEL MAC M1!
)

# Aprende el diccionario de vibraciones
sparse_representation = dict_learner.fit_transform(X_vibracion)
dictionary_atoms = dict_learner.components_

tiempo_total = time.time() - inicio_tiempo
print(f"Motor analizado en {tiempo_total:.2f} segundos.")

# ==========================================
# 4) AUDITORÍA MATEMÁTICA EXACTA
# ==========================================
X_reconstruido = np.dot(sparse_representation, dictionary_atoms)

# Fidelidad (Suma de cuadrados pura, sin raíces)
error_fidelidad = 0.5 * np.sum((X_vibracion - X_reconstruido) ** 2)
# Castigo por activación (Norma L1)
penalizacion_L1 = lambda_param * np.sum(np.abs(sparse_representation))
costo_total = error_fidelidad + penalizacion_L1

print("\n" + "="*60)
print(" DIAGNÓSTICO DEL ALGORITMO (VIBRACIONES INDUSTRIALES)")
print("="*60)
print(f" 1. Fidelidad (Error Puro):         {error_fidelidad:.4f}")
print(f" 2. Castigo de Esparcidad (L1):     {penalizacion_L1:.4f}")
print("-" * 60)
print(f" -> COSTO TOTAL DEL SPARSE CODER:   {costo_total:.4f}")
print("="*60 + "\n")

# ==========================================
# 5) VISUALIZACIÓN DE LAS FIRMAS ACÚSTICAS (LOS "ÁTOMOS")
# ==========================================
print("Generando la radiografía de los átomos mecánicos...")

fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(14, 8))

# Gráfica 1: La señal ruidosa vs La reconstrucción limpia
ax1.plot(X_vibracion.flatten()[:1200], color='lightgray', label='Sensor Bruto (Ruido de fábrica)')
ax1.plot(X_reconstruido.flatten()[:1200], color='tab:red', linewidth=1.5, label='Filtro Sparse (Matemática Pura)')
ax1.set_title('Mantenimiento Predictivo: Limpieza de Señal de Vibración (Primeras 10 ventanas)', fontsize=14, fontweight='bold')
ax1.set_ylabel('Amplitud (Aceleración)', fontsize=12)
ax1.legend(loc="upper right")
ax1.grid(True, linestyle=':', alpha=0.6)



# Gráfica 2: Los Átomos aprendidos
for i in range(4): # Mostramos solo los 4 primeros átomos para no saturar
    desplazamiento = i * 2 # Los separamos visualmente
    ax2.plot(dictionary_atoms[i] + desplazamiento, linewidth=2, label=f'Átomo Mecánico {i}')

ax2.set_title(f'El Vocabulario del Motor (4 de los {n_atoms} Átomos extraídos)', fontsize=14, fontweight='bold')
ax2.set_xlabel('Tiempo dentro de la ventana (milisegundos)', fontsize=12)
ax2.set_ylabel('Firma de resonancia', fontsize=12)
ax2.legend(loc="upper right")
ax2.grid(True, linestyle=':', alpha=0.6)

plt.tight_layout()
plt.show()
```

```python
import scipy.io
import numpy as np
import matplotlib.pyplot as plt
import time
import sys
from sklearn.decomposition import DictionaryLearning
from sklearn.preprocessing import StandardScaler

# ==========================================
# 1) EL MOTOR SANO (LÍNEA BASE)
# ==========================================
print("1. Creando la firma acústica de un motor 100% SANO...")
fs = 12000
t_sano = np.arange(0, 10.0, 1/fs) # 10 segundos de audio
# Un motor sano gira suavemente a 30Hz con un poco de ruido blanco de fondo
senal_sana = 0.2 * np.sin(2 * np.pi * 30 * t_sano) + np.random.normal(0, 0.05, len(t_sano))

longitud_ventana = 200
num_ventanas_sana = len(senal_sana) // longitud_ventana
X_sano = senal_sana[:num_ventanas_sana * longitud_ventana].reshape(num_ventanas_sana, longitud_ventana)

# ==========================================
# 2) EL MOTOR ROTO (TU ARCHIVO CWRU REAL)
# ==========================================
print("2. Cargando tu archivo CWRU real (Motor ROTO)...")
archivo_local = "105.mat" 
try:
    mat_data = scipy.io.loadmat(archivo_local)
except FileNotFoundError:
    print(f"\n[ERROR] No encuentro el '{archivo_local}'.")
    sys.exit()

clave_sensor = [key for key in mat_data.keys() if 'DE_time' in key][0]
senal_rota = mat_data[clave_sensor].flatten()

num_ventanas_rota = len(senal_rota) // longitud_ventana
X_roto = senal_rota[:num_ventanas_rota * longitud_ventana].reshape(num_ventanas_rota, longitud_ventana)

# Para que la suma matemática del error sea justa, igualamos el número de ventanas
ventanas_a_evaluar = min(X_sano.shape[0], X_roto.shape[0])
X_sano = X_sano[:ventanas_a_evaluar]
X_roto = X_roto[:ventanas_a_evaluar]

# ESTANDARIZACIÓN ESTRICTA: Entrenamos el scaler SOLO con el sano
scaler = StandardScaler()
X_sano_scaled = scaler.fit_transform(X_sano)
# Al motor roto se le aplica la misma regla (sin re-entrenar)
X_roto_scaled = scaler.transform(X_roto)

# ==========================================
# 3) ENTRENAMIENTO DEL MODELO (MÁXIMA POTENCIA)
# ==========================================
n_atoms = 30
lambda_param = 1.0

print(f"\n3. Entrenando {n_atoms} átomos con el motor SANO (Usando todos los núcleos)...")
inicio_tiempo = time.time()

dict_learner = DictionaryLearning(
    n_components=n_atoms, 
    alpha=lambda_param, 
    transform_algorithm='lasso_lars', 
    random_state=42,
    max_iter=1000,
    n_jobs=-1  # ¡EL TURBO MULTINÚCLEO!
)

# EL MODELO SOLO VE EL MOTOR SANO
sparse_sano = dict_learner.fit_transform(X_sano_scaled)
dictionary_sano = dict_learner.components_

print(f"   Listo en {time.time() - inicio_tiempo:.2f} segundos.")

# ==========================================
# 4) LA PRUEBA DE FUEGO: RECONSTRUCCIÓN Y ERROR EXACTO
# ==========================================
print("\n4. Obligando al diccionario Sano a reconstruir el motor Roto...")
sparse_roto = dict_learner.transform(X_roto_scaled)

# Reconstrucciones
X_sano_recon = np.dot(sparse_sano, dictionary_sano)
X_roto_recon = np.dot(sparse_roto, dictionary_sano)

# CÁLCULOS EXACTOS DE FIDELIDAD (SIN RAÍCES CUADRADAS)
error_fidelidad_SANO = 0.5 * np.sum((X_sano_scaled - X_sano_recon) ** 2)
error_fidelidad_ROTO = 0.5 * np.sum((X_roto_scaled - X_roto_recon) ** 2)

multiplicador_error = error_fidelidad_ROTO / error_fidelidad_SANO

print("\n" + "="*60)
print(" ALARMA DE MANTENIMIENTO PREDICTIVO (DETECCIÓN DE ANOMALÍAS)")
print("="*60)
print(f" -> Error Exacto en máquina SANA:   {error_fidelidad_SANO:.4f}")
print(f" -> Error Exacto en máquina ROTA:   {error_fidelidad_ROTO:.4f}")
print("-" * 60)
print(f" ¡ALERTA! El error matemático se ha multiplicado por {multiplicador_error:.2f}x")
print("="*60 + "\n")

# ==========================================
# 5) VISUALIZACIÓN DE LA ANOMALÍA
# ==========================================
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(14, 8))

# Panel Sano
ventana_ver = 50
ax1.plot(X_sano_scaled[ventana_ver], color='tab:green', alpha=0.5, linewidth=4, label='Señal Real (Sana)')
ax1.plot(X_sano_recon[ventana_ver], color='black', linestyle='--', linewidth=2, label='Reconstrucción del Diccionario')
ax1.set_title(f'Máquina SANA - Reconstrucción Perfecta (Error: {error_fidelidad_SANO:.0f})', fontsize=14, fontweight='bold')
ax1.legend(loc="upper right")
ax1.grid(True, linestyle=':', alpha=0.6)



# Panel Roto
ax2.plot(X_roto_scaled[ventana_ver], color='tab:red', alpha=0.5, linewidth=4, label='Señal Real (CWRU Rota)')
ax2.plot(X_roto_recon[ventana_ver], color='black', linestyle='--', linewidth=2, label='Reconstrucción Fallida (El diccionario no sabe dibujar esto)')
ax2.set_title(f'Máquina AVERIADA - Explosión Matemática (Error: {error_fidelidad_ROTO:.0f})', fontsize=14, fontweight='bold')
ax2.legend(loc="upper right")
ax2.grid(True, linestyle=':', alpha=0.6)

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import time
from sklearn.decomposition import DictionaryLearning
from sklearn.preprocessing import StandardScaler

# ==========================================
# 1) GENERACIÓN DE IMPACTOS VIOLENTOS (METRÓNOMO)
# ==========================================
print("Generando señal de impactos de alta energía...")
fs = 10000
t = np.arange(0, 1.0, 1/fs)

# Creamos 10 impactos masivos (uno cada 0.1s)
senal = np.zeros_like(t)
for i in range(10):
    inicio = int(i * 0.1 * fs)
    # Impacto muy fuerte que destaca sobre cualquier ruido
    duracion_impacto = np.arange(0, 0.005, 1/fs)
    golpe = 5.0 * np.exp(-500 * duracion_impacto) * np.sin(2 * np.pi * 1500 * duracion_impacto)
    senal[inicio:inicio+len(golpe)] = golpe

# Añadimos un poco de ruido, pero muy poco para no tapar el patrón
senal += np.random.normal(0, 0.1, len(t))

# ==========================================
# 2) MATRIZ Y ESCALADO
# ==========================================
longitud_ventana = 100 # Ventanas más pequeñas para mayor resolución temporal
num_ventanas = len(senal) // longitud_ventana
X = senal[:num_ventanas * longitud_ventana].reshape(num_ventanas, longitud_ventana)

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# ==========================================
# 3) SPARSE CODING (HIPER-SENSIBLE)
# ==========================================
n_atoms = 10
# Bajamos lambda al mínimo para que el algoritmo SE VEA OBLIGADO a pintar
lambda_param = 1.0 

print(f"Procesando con todos los núcleos (Turbo-Sensibilidad)...")
dict_learner = DictionaryLearning(
    n_components=n_atoms, 
    alpha=lambda_param, 
    transform_algorithm='lasso_lars', 
    n_jobs=-1, # Máxima potencia
    random_state=42
)

sparse_rep = dict_learner.fit_transform(X_scaled)

# ==========================================
# 4) VISUALIZACIÓN DE ALTA INTENSIDAD
# ==========================================
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(12, 8), sharex=True)

# Señal original
ax1.plot(np.linspace(0, 1, len(senal)), senal, color='black', alpha=0.7)
ax1.set_title("Señal Física: 10 Impactos Periódicos (Engranaje)", fontweight='bold')

# Mapa de activaciones (Metrónomo)
# Usamos s=100 para que los puntos sean GIGANTES
v, a = np.nonzero(sparse_rep)
fuerzas = np.abs(sparse_rep[v, a])
tiempos = v * (longitud_ventana / fs)

scatter = ax2.scatter(tiempos, a, s=fuerzas * 100, c=fuerzas, cmap='hot', edgecolors='black')
ax2.set_title(f"Partitura de Átomos: Periodicidad Detectada (λ={lambda_param})", fontweight='bold')
ax2.set_ylabel("ID del Átomo")
ax2.set_xlabel("Tiempo (Segundos)")
ax2.set_yticks(range(n_atoms))

# Marcamos las líneas de 0.1s para comprobar la precisión
for i in np.arange(0, 1.1, 0.1):
    ax2.axvline(i, color='blue', linestyle='--', alpha=0.3)

plt.colorbar(scatter, ax=ax2, label="Intensidad de Activación")
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import time
from sklearn.decomposition import DictionaryLearning
from sklearn.preprocessing import StandardScaler

# ==========================================
# 1) SIMULACIÓN PURA DE FÍSICA DE ENGRANAJES
# ==========================================
print("Generando simulación determinista de engranaje (Offline y Segura)...")
fs = 10000  # 10 kHz
t = np.arange(0, 1.0, 1/fs) # 1 segundo exacto de grabación

# Frecuencia de rotación (10 Hz -> 1 vuelta cada 0.1 s. Ahí está la periodicidad del fallo)
f_rot = 10 
# Frecuencia de engranado (supongamos 20 dientes: f_mesh = 10 * 20 = 200 Hz)
f_mesh = f_rot * 20 

# A) El "Canto" normal del engranaje (Senoide pura, Meshing tone)
portadora_mesh = np.sin(2 * np.pi * f_mesh * t)

# B) Los impactos del diente roto (Decaimiento exponencial rítmico)
# Simulamos 10 impactos deterministas y perfectos (uno cada 0.1 s)
impactos_limpios = np.zeros_like(t)
for i in range(1, f_rot + 1):
    # El impacto dura muy poco (0.01 s) y resuena a alta frecuencia (2000 Hz)
    idx_inicio = int((i - 1) * fs * (1/f_rot))
    # Creamos la forma del "golpe" (un 'látigo' amortiguado)
    tiempo_impacto = np.arange(0, 0.01, 1/fs)
    onda_impacto = np.exp(-300 * tiempo_impacto) * np.sin(2 * np.pi * 2000 * tiempo_impacto)
    impactos_limpios[idx_inicio:idx_inicio+len(onda_impacto)] = onda_impacto

# C) El impacto modula la portadora y sumamos ruido caótico de fondo
modulacion_am = (1 + 2.0 * impactos_limpios) # Aumenta la amplitud en el impacto
senal_engranaje = portadora_mesh * modulacion_am + impactos_limpios * 0.5

# D) Sumamos ruido blanco puro de la fábrica (Hacemos que el SC trabaje)
ruido_caotico = np.random.normal(0, 0.4, len(t))
senal_final_cruda = senal_engranaje + ruido_caotico

# ==========================================
# 2) PREPARACIÓN DE LA MATRIZ MATEMÁTICA
# ==========================================
# Cortamos en ventanas de 200 puntos (unos 20 milisegundos de audio)
longitud_ventana = 200
num_ventanas = len(senal_final_cruda) // longitud_ventana

X_vibracion = senal_final_cruda[:num_ventanas * longitud_ventana].reshape(num_ventanas, longitud_ventana)

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X_vibracion)

# ==========================================
# 3) SPARSE CODING (EL TURBO MULTINÚCLEO)
# ==========================================
n_atoms = 8  # <--- TU REGLA: Queremos solo 8 átomos
lambda_param = 5.0  # <--- Castigo alto para forzar una esparcidad bruta y limpiar ruido

print(f"Extrayendo vocabulario de {n_atoms} átomos usando TODOS LOS NÚCLEOS...")
inicio_tiempo = time.time()

dict_learner = DictionaryLearning(
    n_components=n_atoms, 
    alpha=lambda_param, 
    transform_algorithm='lasso_lars', 
    random_state=42,
    max_iter=1000,
    n_jobs=-1  # Siempre usando toda la potencia de tu ordenador
)

sparse_representation = dict_learner.fit_transform(X_scaled)
dictionary_atoms = dict_learner.components_

print(f"Engranaje analizado en {time.time() - inicio_tiempo:.2f} segundos.")

# ==========================================
# 4) AUDITORÍA DE LA FUNCIÓN OBJETIVO EXACTA
# ==========================================
X_reconstruido = np.dot(sparse_representation, dictionary_atoms)

# Matemáticas puras: Costo exacto al cuadrado (Sin aplicar raíces)
error_fidelidad = 0.5 * np.sum((X_scaled - X_reconstruido) ** 2)
penalizacion_L1 = lambda_param * np.sum(np.abs(sparse_representation))
costo_total = error_fidelidad + penalizacion_L1

ceros_exactos = np.sum(np.abs(sparse_representation) < 1e-5)
esparcidad_pct = (ceros_exactos / sparse_representation.size) * 100

print("\n" + "="*60)
print(" DIAGNÓSTICO MATEMÁTICO EXACTO DEL ENGRANAJE")
print("="*60)
print(f" 1. Fidelidad (Error Cuadrático Puro): {error_fidelidad:.4f}")
print(f" 2. Castigo de Esparcidad (L1):        {penalizacion_L1:.4f}")
print("-" * 60)
print(f" -> COSTO TOTAL DEL ALGORITMO:         {costo_total:.4f}")
print(f" -> Nivel de Esparcidad:               {esparcidad_pct:.2f}% de ceros absolutos")
print("="*60 + "\n")

# ==========================================
# 5) VISUALIZACIÓN DEL "METRÓNOMO" PERFECTO
# ==========================================
print("Generando panel de análisis de ingeniería...")

fig = plt.figure(figsize=(14, 10))
gs = fig.add_gridspec(3, 2, height_ratios=[2, 1, 1], width_ratios=[3, 1])

# --- PANEL A: RECONSTRUCCIÓN (Original vs Sparse) ---
ax_recon = fig.add_subplot(gs[0, :])
ax_recon.plot(np.linspace(0, 0.2, 2000), X_scaled[:10].flatten(), color='gray', linewidth=1.0, label='Señal Cruda del Engranaje (Ruido + Diente Roto)')
ax_recon.plot(np.linspace(0, 0.2, 2000), X_reconstruido[:10].flatten(), color='tab:red', linestyle='--', linewidth=2.0, label='Reconstrucción Sparse (El filtro matemático)')
ax_recon.set_title(f'A: Señal Física en el Tiempo (Sensor a 10kHz) - Fidelidad Exacta: {error_fidelidad:.0f}', fontsize=14, fontweight='bold')
ax_recon.set_ylabel('Amplitud', fontsize=12)
ax_recon.set_xlim(0, 0.2)
ax_recon.grid(True, linestyle=':', alpha=0.6)
ax_recon.legend(loc="upper right")
# Marcamos impactos
impact_times = np.arange(0.1, 0.3, 0.1)
for impact_time in impact_times:
    ax_recon.axvline(impact_time, color='black', linestyle='--', linewidth=0.8, alpha=0.5)

# --- PANEL B: LOS 8 ÁTOMOS (Caja de herramientas) ---
ax_atoms = fig.add_subplot(gs[1:, 1])
y_offset = 0
colors = plt.cm.plasma(np.linspace(0, 1, n_atoms))
for i in range(n_atoms):
    ax_atoms.plot(dictionary_atoms[i] + y_offset, color=colors[i], linewidth=2.0)
    ax_atoms.text(210, y_offset, f'Átomo {i}', color=colors[i], fontsize=10, fontweight='bold', verticalalignment='center')
    y_offset += 0.8
ax_atoms.set_title(f'B: Caja de Herramientas (8 Átomos)', fontsize=14, fontweight='bold')
ax_atoms.set_xlim(0, 250)
ax_atoms.set_yticks([])
ax_atoms.set_xticks([])
ax_atoms.grid(True, axis='x', linestyle='--', alpha=0.3)

# --- PANEL C: ESPECTROGRAMAS (Validación física) ---
ax_spec_orig = fig.add_subplot(gs[1, 0])
ax_spec_orig.specgram(senal_final_cruda, NFFT=256, Fs=fs, noverlap=128, cmap='viridis')
ax_spec_orig.set_title('C: Espectrograma Original (Ruido de fábrica)', fontsize=12)
ax_spec_orig.set_ylabel('Frecuencia (Hz)')

ax_spec_recon = fig.add_subplot(gs[2, 0])
# Reconstruimos la señal completa para el espectrograma
senal_recon_completa = X_reconstruido.flatten()
ax_spec_recon.specgram(senal_recon_completa, NFFT=256, Fs=fs, noverlap=128, cmap='viridis')
ax_spec_recon.set_title('D: Espectrograma Sparse (Solo impactos)', fontsize=12)
ax_spec_recon.set_xlabel('Tiempo (Segundos)')
ax_spec_recon.set_ylabel('Frecuencia (Hz)')

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import time
from sklearn.decomposition import DictionaryLearning
from sklearn.preprocessing import StandardScaler

# ==========================================
# 1) SIMULACIÓN FÍSICA: SANO VS ROTO
# ==========================================
print("1. Creando las físicas de los engranajes (Sano y Roto)...")
fs = 10000  # 10 kHz
t = np.arange(0, 1.0, 1/fs) # 1 segundo exacto

f_rot = 10 # 10 Hz -> 1 vuelta cada 0.1 s.
f_mesh = 200 # Frecuencia de engranado normal

# Engranaje SANO (Onda pura + ruido natural de fondo)
senal_sana = np.sin(2 * np.pi * f_mesh * t) + np.random.normal(0, 0.3, len(t))

# Engranaje ROTO (Diente partido que golpea cada 0.1s)
impactos = np.zeros_like(t)
for i in range(11): # Un impacto cada 0.1s
    idx = int(i * fs * 0.1)
    if idx < len(t):
        tiempo_imp = np.arange(0, 0.01, 1/fs)
        onda_imp = 4.0 * np.exp(-400 * tiempo_imp) * np.sin(2 * np.pi * 2500 * tiempo_imp)
        impactos[idx:idx+len(onda_imp)] = onda_imp[:len(t)-idx]

senal_rota = np.sin(2 * np.pi * f_mesh * t) * (1 + 0.5 * impactos) + impactos + np.random.normal(0, 0.3, len(t))

# ==========================================
# 2) PREPARACIÓN DE LAS MATRICES (Ventanas de 0.02s)
# ==========================================
longitud_ventana = 200 # 20 milisegundos por ventana
num_ventanas = len(t) // longitud_ventana

X_sano = senal_sana[:num_ventanas * longitud_ventana].reshape(num_ventanas, longitud_ventana)
X_roto = senal_rota[:num_ventanas * longitud_ventana].reshape(num_ventanas, longitud_ventana)

# ESTANDARIZACIÓN ESTRICTA: El scaler solo aprende de la máquina sana
scaler = StandardScaler()
X_sano_scaled = scaler.fit_transform(X_sano)
X_roto_scaled = scaler.transform(X_roto) # Se aplica la misma regla al roto

# ==========================================
# 3) APRENDIZAJE DEL DICCIONARIO SANO (MÁXIMA POTENCIA)
# ==========================================
n_atoms = 8
lambda_param = 1.0 

print(f"\n2. Entrenando {n_atoms} átomos perfectos usando TODOS LOS NÚCLEOS...")
inicio = time.time()

dict_learner = DictionaryLearning(
    n_components=n_atoms, 
    alpha=lambda_param, 
    transform_algorithm='lasso_lars', 
    random_state=42,
    max_iter=1000,
    n_jobs=-1  # ¡Todos los núcleos a la vez!
)

# EL MODELO SOLO VE EL ENGRANAJE SANO
dict_learner.fit(X_sano_scaled)
dictionary_sano = dict_learner.components_

print(f"   Diccionario sano completado en {time.time() - inicio:.2f} segundos.")

# ==========================================
# 4) RECONSTRUCCIÓN Y CÁLCULO DE ERROR EXACTO
# ==========================================
print("\n3. Obligando a las herramientas sanas a reconstruir la máquina rota...")
alpha_sano = dict_learner.transform(X_sano_scaled)
alpha_roto = dict_learner.transform(X_roto_scaled)

X_sano_recon = alpha_sano @ dictionary_sano
X_roto_recon = alpha_roto @ dictionary_sano

# CÁLCULO EXACTO POR VENTANA (Matemática pura sin raíces)
errores_ventanas_sano = 0.5 * np.sum((X_sano_scaled - X_sano_recon)**2, axis=1)
errores_ventanas_roto = 0.5 * np.sum((X_roto_scaled - X_roto_recon)**2, axis=1)

print("\n" + "="*60)
print(" AUDITORÍA DE ERROR EXACTO POR VENTANA (0.02s cada una)")
print(" Copia este bloque en el chat para analizarlo juntos")
print("="*60)
print("Ventana | Error SANO | Error ROTO (¡Busca la explosión!)")
print("-" * 60)
for i in range(13): # Mostramos las primeras 13 ventanas (0.26 segundos)
    print(f" Vent.{i:2d} |  {errores_ventanas_sano[i]:7.2f}   |  {errores_ventanas_roto[i]:8.2f}")
print("="*60 + "\n")

# ==========================================
# 5) VISUALIZACIÓN MULTIPANEL
# ==========================================
print("Generando el panel de ingeniería...")

fig = plt.figure(figsize=(15, 12))
gs = fig.add_gridspec(4, 2, height_ratios=[1, 1, 1, 1.5], width_ratios=[3, 1])

# Panel 1: Engranaje Sano
ax_sano = fig.add_subplot(gs[0, 0])
ax_sano.plot(np.linspace(0, 0.3, 3000), X_sano_scaled[:15].flatten(), color='gray', alpha=0.5, label='Física Real Sana')
ax_sano.plot(np.linspace(0, 0.3, 3000), X_sano_recon[:15].flatten(), color='tab:green', linestyle='--', label='Reconstrucción Perfecta')
ax_sano.set_title('1. Máquina Sana: El modelo calca la onda sin esfuerzo', fontweight='bold')
ax_sano.legend(loc='upper right')
ax_sano.set_xlim(0, 0.3)

# Panel 2: Los 8 Átomos Sanos
ax_atoms = fig.add_subplot(gs[0:2, 1])
y_offset = 0
for i in range(n_atoms):
    ax_atoms.plot(dictionary_sano[i] + y_offset, color='tab:blue')
    ax_atoms.text(210, y_offset, f'Átomo {i}', verticalalignment='center')
    y_offset += 1.5
ax_atoms.set_title('2. Los 8 Átomos (Solo saben\ndibujar curvas suaves)', fontweight='bold')
ax_atoms.set_yticks([])
ax_atoms.set_xticks([])



# Panel 3: Engranaje Roto
ax_roto = fig.add_subplot(gs[1, 0])
ax_roto.plot(np.linspace(0, 0.3, 3000), X_roto_scaled[:15].flatten(), color='gray', alpha=0.5, label='Física Real Rota (Con Impactos)')
ax_roto.plot(np.linspace(0, 0.3, 3000), X_roto_recon[:15].flatten(), color='tab:red', linestyle='--', label='Reconstrucción Fallida')
ax_roto.set_title('3. Máquina Rota: Las herramientas sanas FRACASAN en los golpes', fontweight='bold')
ax_roto.legend(loc='upper right')
ax_roto.set_xlim(0, 0.3)
for impact_t in [0.0, 0.1, 0.2]:
    ax_roto.axvline(impact_t, color='black', linestyle=':')

# Panel 4: Espectrograma del Engranaje ROTO (La señal completa que vamos a reconstruir)
ax_spec_orig = fig.add_subplot(gs[2, :])
ax_spec_orig.specgram(senal_rota, NFFT=256, Fs=fs, noverlap=128, cmap='viridis')
ax_spec_orig.set_title('4. Espectrograma de la Señal Rota Original (Mucho ruido)', fontweight='bold')
ax_spec_orig.set_ylabel('Frecuencia (Hz)')
ax_spec_orig.set_xlim(0, 1.0)

# Panel 5: Espectrograma de la RECONSTRUCCIÓN
ax_spec_recon = fig.add_subplot(gs[3, :])
ax_spec_recon.specgram(X_roto_recon.flatten(), NFFT=256, Fs=fs, noverlap=128, cmap='magma')
ax_spec_recon.set_title('5. Espectrograma de la Reconstrucción (Las "cicatrices" matemáticas revelan los impactos cada 0.1s)', fontweight='bold')
ax_spec_recon.set_xlabel('Tiempo (Segundos)')
ax_spec_recon.set_ylabel('Frecuencia (Hz)')
ax_spec_recon.set_xlim(0, 1.0)



plt.tight_layout()
plt.show()
```
